```python
# import module ที่ต้องใช้
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# StatsModels
!pip install statsmodels --quiet
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Download the file
!wget https://donlapark.pages.dev/229351/data/Auto.csv
```

```python
# อ่านไฟล์ csv ก่อนเพื่อหา missing values
auto = pd.read_csv('Auto.csv', na_values=["?"])

# ลบแถวที่มี missing values
auto = auto.dropna()

auto = auto.drop(['name', 'origin'],axis=1)

auto.head()
```

# Linear model diagnosis

```python
# simple linear regression on Auto dataset
auto_lr = smf.ols('mpg ~ horsepower ', data=auto).fit()
print(auto_lr.summary())
```

### คำนวณ Residuals $y_i-\hat{y}_i$

```python
#Fitted values
auto_fitted = auto_lr.predict(auto)

#Residuals
auto_res = auto['mpg']-auto_fitted

print(auto_res)
```

### หรือสามารถดึง predictions และ residuals จากโมเดลได้โดยตรง

```python
# Fitted values
auto_fitted_ = auto_lr.fittedvalues

# Residuals
auto_res_ = auto_lr.resid
```

### 1. ทดสอบ non-linearity

```python
plt.scatter(auto_fitted, auto_res)
plt.title('Residuals vs Fitted')
plt.xlabel('Fitted values')
plt.ylabel('Residuals');
```

```python
plt.scatter(auto_lr2.fittedvalues, auto_lr2.resid)
plt.title('Residuals vs Fitted')
plt.xlabel('Fitted values')
plt.ylabel('Residuals');
```

### 2. ทดสอบ independence of errors
2.1 ตรวจสอบโดยการ plot time vs residuals (ถ้าสามารถทำได้)

2.2 ตรวจสอบด้วย Durbin-Wutson test  

H0: There is no correlation among the residuals.  
H1: The residuals are correlated.

The Durbin-Watson statistic ranges in value from 0
to 4. A value near 2 indicates non-correlation; a value toward 0 indicates positive
correlation; a value toward 4 indicates negative correlation.

Reject H0 if d < dL  
Accept H0 if d > dU

Durbin-Watson table: https://www.real-statistics.com/statistics-tables/durbin-watson-table/

```python
auto_lr2 = smf.ols('mpg ~ horsepower+I(horsepower**2)', data=auto).fit()
print(auto_lr2.summary())
```

### ดูตาราง Durbin-Watson Alpha = .05 n = 350 และ k = 2

ได้ค่า dL = 1.813 dU = 1.836

ดังนั้น Reject H0 -- มี correlation ระหว่าง residuals

### 3. ทดสอบ constant variance

### ใช้ Breush-Pagan test


H0: The residuals are distributed with equal variance  
H1: The residuals are not distributed with equal variance


```python
import statsmodels.stats.api as sms

name = ["Lagrange multiplier statistic", "p-value", "f-value", "f p-value"]
test = sms.het_breuschpagan(auto_lr2.resid, auto_lr2.model.exog)
pd.DataFrame({'Name':name, 'value':test})
```

### p-value < 0.05 ดังนั้น accept H1

```python
auto_lr3 = smf.ols('I(np.log(mpg)) ~ horsepower+I(horsepower**2)', data=auto).fit()
print(auto_lr3.summary())
```

```python
plt.scatter(auto_lr3.fittedvalues, auto_lr3.resid)
plt.title('Residuals vs Fitted')
plt.xlabel('Fitted values')
plt.ylabel('Residuals');
```

### ลอง Breush-Pagan test อีกครั้ง

```python
name = ["Lagrange multiplier statistic", "p-value", "f-value", "f p-value"]
test = sms.het_breuschpagan(auto_lr3.resid, auto_lr3.model.exog)
pd.DataFrame({'Name':name, 'value':test})
```

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Load the dataset
carseats = pd.read_csv('Carseats.csv')
```

```python
# Fit the multiple linear regression model
# The formula includes all predictors against the 'Sales' variable.
# statsmodels automatically handles the conversion of categorical variables (like ShelveLoc) into dummy variables.
all_predictors = '+'.join(carseats.columns.drop('Sales'))
formula = 'Sales ~ ' + all_predictors
model = smf.ols(formula=formula, data=carseats)
results = model.fit()
```

```python
# Get the coefficients for ShelveLoc
shelveloc_coeffs = results.params.filter(like='ShelveLoc')
medium_coeff = shelveloc_coeffs.get('ShelveLoc[T.Medium]', 0)
good_coeff = shelveloc_coeffs.get('ShelveLoc[T.Good]', 0)
```

# 1.สมมติว่าตัวแปรอื่นมีค่าเท่ากัน โดยเฉลี่ยแล้ว ยอดขายของเบาะรถที่วางบนชั้นที่มีคุณภาพปานกลาง (ShelveLoc (Medium)) กับชั้นที่มีคุณภาพต่ำ (ShelveLoc (Bad)) ต่างกันอย่างไร ต่างกันเท่าไหร่

```python
diff_medium_vs_bad = medium_coeff
print(f" Difference Medium vs. Bad: {diff_medium_vs_bad:.4f}")
```

# 2.สมมติว่าตัวแปรอื่นมีค่าเท่ากัน โดยเฉลี่ยแล้ว ยอดขายของเบาะรถที่วางบนชั้นที่มีคุณภาพดี (ShelveLoc (Good)) กับชั้นที่มีคุณภาพปานกลาง (ShelveLoc (Medium)) ต่างกันอย่างไร ต่างกันเท่าไหร่

```python
diff_good_vs_medium = good_coeff - medium_coeff
print(f" Difference Good vs. Medium: {diff_good_vs_medium:.4f}")
```

```python
# Get fitted values and residuals
fitted_vals = results.fittedvalues
residuals = results.resid
studentized_residuals = results.get_influence().resid_studentized_internal
```

```python
# Create Residual Plot (Residuals vs. Fitted Values)
plt.figure(figsize=(10, 6))
sns.residplot(x=fitted_vals, y=residuals, lowess=True,
              scatter_kws={'alpha': 0.5},
              line_kws={'color': 'red', 'lw': 2, 'linestyle': '--'})
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.grid(True)
# plt.savefig('residual_plot.png') # In a real scenario, we'd save the plot
print("Residual Plot Generated.")

# Create Influence Plot
fig, ax = plt.subplots(figsize=(12, 8))
sm.graphics.influence_plot(results, ax=ax, criterion="cooks")
plt.title('Influence Plot')
plt.grid(True)
# plt.savefig('influence_plot.png') # In a real scenario, we'd save the plot
print("Influence Plot Generated.")

# Independence of Errors (Durbin-Watson)
durbin_watson_stat = sm.stats.stattools.durbin_watson(results.resid)

# 3.6 Multicollinearity (VIF)
# Select only numeric predictors for VIF calculation
numeric_predictors = carseats.select_dtypes(include=['number']).drop(columns='Sales')
# Add a constant for the intercept
X = sm.add_constant(numeric_predictors)
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Print results for review
print("\n--- Regression Results Summary ---")
print(results.summary())
```

# 3.จงตรวจสอบว่าข้อมูลที่ใช้มีความเหมาะสมกับการสร้าง linear regression หรือไม่ โดยพิจารณาสิ่งต่างๆเหล่านี้พร้อมกับให้เหตุผล

# 3.1 Non-linearity

จากกราฟ "Residuals vs. Fitted Values" จะเห็นว่าการกระจายตัวของจุดข้อมูล (residuals) เป็นไปอย่างสุ่มรอบเส้นศูนย์ และเส้นแนวโน้มสีแดง (LOWESS line) ค่อนข้างเรียบและวิ่งขนานไปกับเส้นศูนย์ ซึ่งบ่งชี้ว่าความสัมพันธ์ระหว่างตัวแปรอิสระและตัวแปรตามมีความเป็นเชิงเส้น (Linear Relationship)

# 3.2 Independence of errors

```python
print(f" Durbin-Watson statistic: {durbin_watson_stat:.4f}")
```

ค่าสถิติ Durbin-Watson ที่คำนวณได้คือ 2.08 ซึ่งมีค่าใกล้เคียง 2  ค่านี้ใช้ตรวจสอบว่าค่าความคลาดเคลื่อน (errors) ของข้อมูลที่อยู่ติดกันมีความสัมพันธ์กันหรือไม่ (Autocorrelation) เมื่อค่าเข้าใกล้ 2 หมายความว่าค่าความคลาดเคลื่อนเป็นอิสระต่อกัน ซึ่งเป็นไปตามข้อตกลงเบื้องต้น

# 3.3 Constant variance

พิจารณาจากกราฟ "Residuals vs. Fitted Values" จะเห็นว่าความกว้างของการกระจายตัวของจุดในแนวตั้งมีความสม่ำเสมอและคงที่ตลอดช่วงของค่าที่ทำนาย (Fitted Values) ไม่ปรากฏลักษณะของรูปกรวย (cone shape) ซึ่งแสดงว่าข้อมูลมีคุณสมบัติความแปรปรวนคงที่ (Homoscedasticity)

# 3.4 Outliers

จากกราฟ "Influence Plot" ค่าผิดปกติจะถูกพิจารณาจากแกน Y (Studentized Residuals) จากกราฟจะเห็นว่ามีบางจุดที่เกิน +2 และ -2 แต่ไม่มีจุดใดที่มีค่าเกินเกณฑ์ +/- 3 อย่างชัดเจน จึงสรุปได้ว่าไม่มีค่าผิดปกติที่ส่งผลกระทบต่อโมเดล

# 3.5 Leverage points

Leverage Points คือจุดข้อมูลที่มีค่าของตัวแปรอิสระแตกต่างจากข้อมูลส่วนใหญ่อย่างชัดเจน ซึ่งดูได้จากแกน X ใน "Influence Plot" จากกราฟไม่พบจุดข้อมูลใดที่อยู่ห่างไปทางขวาของกราฟมากผิดปกติ แสดงว่าไม่มีจุดข้อมูลใดมีอิทธิพลสูง (High Leverage) ต่อการสร้างเส้นถดถอย
